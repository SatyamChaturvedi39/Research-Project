```python
def load_nba_data():
    teams_df = pd.read_csv('C:\\MSAIML\\research\\Research-Project\\data\\raw\\24_25_teams.csv')
    
    players_df = pd.read_csv('C:\\MSAIML\\research\\Research-Project\\data\\raw\\nba_2024_25_totals.csv')
    
    players_df['PPG'] = players_df['PTS'] / players_df['G']  # Points per game
    players_df['APG'] = players_df['AST'] / players_df['G']  # Assists per game
    players_df['RPG'] = players_df['TRB'] / players_df['G']  # Rebounds per game
    players_df['MPG'] = players_df['MP'] / players_df['G']   # Minutes per game
    
    # Calculate variance for each stat (based on league average variance)
    players_df['PPG_var'] = (players_df['PPG'] * 0.15) ** 2  # 15% coefficient of variation
    players_df['APG_var'] = (players_df['APG'] * 0.20) ** 2  # 20% coefficient of variation
    players_df['RPG_var'] = (players_df['RPG'] * 0.15) ** 2  # 15% coefficient of variation
    
    print(f"✓ Loaded {len(teams_df)} teams")
    print(f"✓ Loaded {len(players_df)} players")
    
    return teams_df, players_df
```

```python
load_nba_data()
```

```python
def get_team_stats(teams_df, team_name):
    team_row = teams_df[teams_df['Team'].str.contains(team_name, case=False, na=False)]
    
    if team_row.empty:
        raise ValueError(f"Team '{team_name}' not found!")
    
    team = team_row.iloc[0]
    
    return {
        'name': team['Team'],
        'wins': int(team['W']),
        'losses': int(team['L']),
        'offensive_rating': float(team['ORtg']),  
        'defensive_rating': float(team['DRtg']),  
        'pace': float(team['Pace']),
        'net_rating': float(team['NRtg'])
    }
```
